```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    return 3*x**2 - 4*x + 5
```

```python
f(5.0)
```

```python
xs = np.arange(-5, 5, 0.25)
ys = f(xs)
```

```python
ys
```

```python
plt.plot(xs, ys)
```

```python
h = 0.001
x = 3.0
f(x+h)
```

```python
(f(x+h) - f(x))/h
```

```python
def g(x):
    return 6*x - 4
```

```python
ys2 = g(xs)
```

```python
ys2
```

```python
plt.plot(xs, ys)
plt.plot(xs, ys2)
```

```python
a = 2.0
b = -3.0
c = 10.0
d = a*b+c
print(d)
```

```python
h = 0.0001

a = 2.0
b = -3.0
c = 10.0
d1 = a*b+c
a+=h
d2 = a*b + c
```

```python
print('d1', d1)
print('d2', d2)

print('slope', (d2-d1)/h)
```
